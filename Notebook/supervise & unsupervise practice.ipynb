```python
import numpy as np
from pylab import *
```

```python
np.random.seed
pageSpeeds = np.random.normal(0,4,100)
purchaseAmount = np.random.normal(3,5,100)/pageSpeeds

scatter(pageSpeeds,purchaseAmount)
```

```python
trainX = pageSpeeds[:80]
testX = pageSpeeds[80:]

trainY = purchaseAmount[:80]
testY = purchaseAmount[80:]
```

```python
trainX
```

```python
scatter(trainX,trainY)
```

```python
scatter(testX, testY)
```

```python
x = np.array(trainX)
y = np.array(trainY)

p4 = np.poly1d(np.polyfit(x,y,12))
```

```python

import matplotlib.pyplot  as plt

xp = np.linspace(0,7,100)
axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([-50,200])
plt.scatter(x,y)
plt.plot(xp,p4(xp), c='r')
plt.show()
```

```python
testx = np.array(testX)
testy = np.array(testY)



axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([-50,200])
plt.scatter(testx,testy)
plt.plot(xp,p4(xp), c='r')
plt.show()
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(np.array(trainY), p4(np.array(trainX)))
print(r2)
```
