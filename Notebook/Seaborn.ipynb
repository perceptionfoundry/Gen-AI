# DISTRIBUTION BLOCK

```python
import seaborn as sns

%matplotlib inline

tips = sns.load_dataset("tips")

tips.head()
```

```python
sns.displot(tips["total_bill"], kde = True)
```

```python
sns.displot(tips["total_bill"], kde = False)
```

```python
sns.displot(tips["total_bill"], kde = True, bins= 30)
```

```python
sns.jointplot(x="total_bill",y="tip", data = tips)
```

```python
sns.jointplot(x="total_bill",y="tip", data = tips, kind = "hex")
```

```python
sns.jointplot(x="total_bill",y="tip", data = tips, kind= "reg")
```

```python
sns.jointplot(x="total_bill",y="tip", data = tips, kind= "kde")
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue="sex")
```

```python
sns.pairplot(tips, hue="sex", palette="coolwarm")
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips["total_bill"])
```

# MATRIX PLOTS

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
tc = tips.corr()  # create matrix
tc.head()
```

```python
sns.heatmap(tc)
```

```python
sns.heatmap(tc, annot=True, cmap="coolwarm")
```

```python
flight_mat = flights.pivot_table(index="month",columns="year",values="passengers")
flight_mat
```

```python
sns.heatmap(flight_mat,cmap="magma",linecolor = "white", linewidths= 2)
```

```python
sns.clustermap(flight_mat,cmap= "coolwarm", standard_scale=1)
```

# Regression Plot

```python
sns.lmplot(x="total_bill",y="tip",data= tips, hue="sex",markers=["o",'v'],scatter_kws={'s': 50})
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,col="sex",row="time")
```

```python
sns.lmplot(x="total_bill",y="tip",data= tips, hue="sex",col="day",aspect=0.6, height=6)
```

# GRID

```python
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
```

```python
g.map_diag(sns.histplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
f = sns.FacetGrid(tips, col='time',row="smoker")
f.map(sns.histplot,"total_bill")
```
