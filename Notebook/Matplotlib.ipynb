```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#FUNCTIONAL
plt.plot(x,y)
plt.xlabel(' x  label')
plt.ylabel('y label')
plt.title('title')
plt.show()
```

```python


plt.subplot(1,2,1)
plt.plot(x,y, 'red')

plt.subplot(1,2,2)
plt.plot(x,y, 'blue')
```

```python
fig = plt.figure()
axe = fig.add_axes([0.1,0.1,0.8,0.8])
axe.plot(x,y)
axe.set_xlabel("X")
axe.set_ylabel("Y")
axe.set_title("AXE")
```

```python
fig = plt.figure()
a1 = fig.add_axes([0.1,0.1,0.8,0.8])
a1.set_title ("big")
a2 = fig.add_axes([0.2, 0.5, 0.4,0.3])
a2.set_title("small")

a1.plot(x,y)
a2.plot(y,x)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[1].plot(x,y)
axes[0].plot(y,x)
plt.tight_layout()
```

```python
fig1, axes1 = plt.subplots(nrows = 2, ncols=1, figsize=(8,2), dpi = 200)

axes1[0].plot(x,y)
axes1[1].plot(y,x)
plt.tight_layout()
```

```python
fig1
```

```python
fig1.savefig("graph.png")
```

```python
fig2 = plt.figure()
axis2 = fig2.add_axes([0,0,1,1])
axis2.plot(x,x**2 , label = "X square")
axis2.plot(x,x**3, label = "X cube")

axis2.legend(loc= 10)
```

# PLOT Appearance

```python
fig_app = plt.figure()


ax_app  = fig_app.add_axes([0,0,1,1])
ax_app.plot(x,y, color= 'purple', lw = 0.5, linestyle= '-',
            marker = "o", markersize = 20,
            markerfacecolor = 'yellow', markeredgewidth = 3, markeredgecolor = 'green' )
```
